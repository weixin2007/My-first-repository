```python
def missing_values_summary(df):
    """
    Returns each column name with the number of missing values

    Parameters:
    df (pd.DataFrame): Input dataframe

    Returns:
    pd.Series: Column names with missing value counts
    """
    missing_counts = df.isna().sum()
    return missing_counts[missing_counts > 0]  # Return only columns with missing values

# Usage
missing_values = missing_values_summary(df)
print("Columns with missing values:")
print(missing_values)
```

```python
# Group similar products
df['product_group'] = df['category'] + '_' + df['brand']
```

```python
# Add revenue to data framework
df['revenue'] = df['current_price'] * df['stock_quantity']
# Rank products within each group
grouped = df.groupby('product_group')['revenue'].sum().reset_index()
grouped['rank'] = grouped['revenue'].rank(ascending=False)

```

```python
# Identify Low-Ranked Products
low_ranked = grouped[grouped['rank'] > grouped.groupby('product_group')['rank'].transform(lambda x: x.quantile(0.9))]
```

```python
# Step 5: Simulate removal
remaining_products = df.drop(low_ranked.index)

total_revenue_before = df['revenue'].sum()
total_revenue_after = remaining_products['revenue'].sum()

# If Revenue loss=0: those low-ranked products were not adding new sales
# If Revenue loss>0: those low-ranked products were adding new sales
print("Revenue loss:", total_revenue_before - total_revenue_after)

```

```python
# Generate table 1 (Properties)

table1_data = []

for col in df.columns:
    # --- Basic Properties ---
    dtype = df[col].dtype
    unique_vals = df[col].nunique()
    null_vals = df[col].isnull().sum()

    # --- Placeholders (Default values) ---
    min_val = "-"
    max_val = "-"
    range_val = "-"
    outlier_status = "No"

    # Calculate 'Top Value' (Mode) safely
    try:
        top_val = df[col].mode()[0]
        # Round if it's a number
        if isinstance(top_val, (int, float)):
            top_val = round(top_val, 2)
    except:
        top_val = "-"

    # --- Logic for Numeric Columns ---
    if pd.api.types.is_numeric_dtype(df[col]):
        # Rounding to 2 decimal places for neatness
        min_val = round(df[col].min(), 2)
        max_val = round(df[col].max(), 2)
        range_val = f"{min_val} - {max_val}"

        # [cite_start]Check for Outliers (Z-score > 3) [cite: 163]
        clean_col = df[col].dropna()
        if len(clean_col) > 0:
            z_scores = np.abs(stats.zscore(clean_col))
            if (z_scores > 3).any():
                outlier_status = "Yes"

    # --- Append to List ---
    table1_data.append({
        "Variable": col,
        "Types of Data": "Numeric" if pd.api.types.is_numeric_dtype(df[col]) else "Categorical",
        "Data Types": str(dtype),
        "Measurement Level": "-",  # MANUAL ENTRY REQUIRED in Word (e.g., Ratio, Nominal)
        "Units": "-",              # MANUAL ENTRY REQUIRED in Word (e.g., USD, Years)
        "Range": range_val,
        "Min Value": min_val,
        "Top Value": top_val,
        "Unique Values": unique_vals,
        "Null Values": null_vals,
        "Outliers": outlier_status
    })

table1_df = pd.DataFrame(table1_data)
print("--- Table 1: Data Properties ---")
display(table1_df)

# Generate table 2 (Statistics)

numeric_cols = df.select_dtypes(include=[np.number]).columns
table2_data = []

for col in numeric_cols:
    # Frequency & Completeness
    count = df[col].count()
    completeness = f"{(count / len(df)) * 100:.1f}%"

    # Central Tendency & Dispersion
    mean = df[col].mean()
    median = df[col].median()
    std_dev = df[col].std()
    variance = df[col].var()
    skew = df[col].skew()
    kurt = df[col].kurt()

    # Mode (Most frequent)
    try:
        mode = df[col].mode()[0]
    except:
        mode = "-"

    # Percentiles (25th, 50th, 75th)
    p25 = df[col].quantile(0.25)
    p50 = df[col].quantile(0.50)
    p75 = df[col].quantile(0.75)
    percentiles = f"25th:{p25:.2f}, 50th:{p50:.2f}, 75th:{p75:.2f}"

    # --- Append to List ---
    table2_data.append({
        "Variable": col,
        "Frequency": count,
        "Percentile (25th, 50th, 75th)": percentiles,
        "Data Completeness": completeness,
        "Mean": round(mean, 2),
        "Median": round(median, 2),
        "Mode": round(mode, 2) if isinstance(mode, (int, float)) else mode,
        "Standard Deviation": round(std_dev, 2),
        "Variance": round(variance, 2),
        "Skewness": round(skew, 2),
        "Kurtosis": round(kurt, 2)
    })

table2_df = pd.DataFrame(table2_data)
print("\n--- Table 2: Statistics ---")
display(table2_df)

# ==========================================
# STEP 4: EXPORT TO CSV (FOR REPORT)
# ==========================================
table1_df.to_csv('Final_Table1.csv', index=False)
table2_df.to_csv('Final_Table2.csv', index=False)
print("\n[SUCCESS] 'Final_Table1.csv' and 'Final_Table2.csv' have been saved. Please download them from the Files tab.")
```
